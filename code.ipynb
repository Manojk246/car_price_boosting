```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv(r'C:\AI workforce\Car-price Boosting\CAR DETAILS FROM CAR DEKHO.csv')
df.head(10)
```

```python
df = df.drop_duplicates()
df
```

```python
df.describe()
```

```python
df.hist(figsize=(12, 8), bins=30)
plt.suptitle("Histograms of Numerical Features", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True,  linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features", fontsize=14, pad=12)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="year", y="selling_price", data=df, alpha=0.6, color="navy")
plt.title("Selling Price vs Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['selling_price'], bins=30, kde=True, color="steelblue")
plt.title("Selling Price Distribution")
plt.xlabel("Selling Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df['selling_price_log'] = np.log1p(df['selling_price'])  
sns.histplot(df['selling_price_log'], bins=30, kde=True, color="green")
plt.title("Selling Price (Log Transformed)")
plt.show()
```

```python
sns.histplot(df['year'], bins=30, kde=True, color="steelblue")
plt.title("Year distribution")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='year', data=df, order=sorted(df['year'].unique()))
plt.title('Year distribution (counts)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['km_driven_log'] = np.log1p(df['km_driven'])  
sns.histplot(df['km_driven_log'], bins=30, kde=True, color="green")
plt.title("km_driven (Log Transformed)")
plt.show()
```
